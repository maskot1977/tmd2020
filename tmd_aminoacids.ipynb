# RDKitによるモルオブジェクトの生成

```
import numpy as np
from rdkit import Chem
mols = np.array([Chem.MolFromSmiles(smile) for smile in df['SMILES']])
```

```
def get_list(colname, value):
    return [i for i, b in enumerate(df[colname] == value) if b]
```

```
from rdkit.Chem import Draw
list_ = get_list('Class', 'Common amino acids')
Draw.MolsToGridImage(mols[list_], molsPerRow=5, useSVG=False, 
                          legends=[n for n in df['Name'].iloc[list_]])
```

# RDKit descriptor

```
from rdkit.Chem import Descriptors
names = [x[0] for x in Descriptors._descList if x[0]]
names
```

```
from rdkit.ML.Descriptors import MoleculeDescriptors
calc = MoleculeDescriptors.MolecularDescriptorCalculator(names)
```

```
desc = pd.DataFrame([calc.CalcDescriptors(mol) for mol in mols],
             columns=names)
desc
```

```
import matplotlib.pyplot as plt
for name in names:
    plt.title(name)
    desc[name].hist(bins=20)
    plt.show()
```

# PCA

```
from sklearn.decomposition import PCA #主成分分析器

#主成分分析の実行
col = []
for i in range(len(desc.columns)):
    try:
        pca = PCA()
        pca.fit(desc.iloc[:, col + [i]])
        col.append(i)
    except:
        pass

desc = desc.iloc[:, col]
```

```
desc = desc.apply(lambda x: (x-x.mean())/x.std(), axis=0)
```

```
desc = desc.dropna(axis=1)
```

```
from sklearn.decomposition import PCA #主成分分析器

#主成分分析の実行
pca = PCA()
pca.fit(desc)
# データを主成分空間に写像 = 次元圧縮
feature = pca.transform(desc)
```

```
# 累積寄与率を図示する
import matplotlib.ticker as ticker
import numpy as np
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
plt.plot([0] + list( np.cumsum(pca.explained_variance_ratio_)), "-o")
plt.xlabel("Number of principal components")
plt.ylabel("Cumulative contribution ratio")
plt.grid()
plt.show()
```

```
# 第一主成分と第二主成分でプロットする
plt.figure(figsize=(12, 12))

cl = 'Common amino acids'
list_ = get_list('Class', cl)
plt.scatter(feature[list_, 0], feature[list_, 1], label=cl) #, alpha=0.9)
for x, y, name in zip(feature[list_, 0], feature[list_, 1], df.iloc[list_, 1]):
    plt.text(x, y, name)

cl = 'D-amino acids'
list_ = get_list('Class', cl)
plt.scatter(feature[list_, 0], feature[list_, 1], label=cl, marker='+') #, alpha=0.9)

cl = 'Other amino acids'
list_ = get_list('Class', cl)
plt.scatter(feature[list_, 0], feature[list_, 1], label=cl, marker='x') #, alpha=0.9)

cl = 'Biogenic amines'
list_ = get_list('Class', cl)
plt.scatter(feature[list_, 0], feature[list_, 1], label=cl, marker='^') #, alpha=0.9)

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid()
plt.legend()
plt.show()

plt.figure(figsize=(12, 12))

cl = 'Common amino acids'
list_ = get_list('Class', cl)
plt.scatter(feature[list_, 2], feature[list_, 3], label=cl) #, alpha=0.9)
for x, y, name in zip(feature[list_, 2], feature[list_, 3], df.iloc[list_, 1]):
    plt.text(x, y, name)

cl = 'D-amino acids'
list_ = get_list('Class', cl)
plt.scatter(feature[list_, 2], feature[list_, 3], label=cl, marker='+') #, alpha=0.9)

cl = 'Other amino acids'
list_ = get_list('Class', cl)
plt.scatter(feature[list_, 2], feature[list_, 3], label=cl, marker='x') #, alpha=0.9)

cl = 'Biogenic amines'
list_ = get_list('Class', cl)
plt.scatter(feature[list_, 2], feature[list_, 3], label=cl, marker='^') #, alpha=0.9)


plt.xlabel('PC3')
plt.ylabel('PC4')
plt.grid()
plt.legend()
plt.show()
```

# Morgan fingerprint

```
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem

morgans = [] # 算出した MF を格納 
no_error = [] # エラーなく計算できたIDを格納
for idx, smile in enumerate(df['SMILES']):
    mol = Chem.MolFromSmiles(smile)
    try:
        fp = AllChem.GetMorganFingerprint(mol, 2) # MF を算出
    except:
        continue
    no_error.append(idx)
    morgans.append(fp.GetNonzeroElements())

morgan_keys = [] # 全化合物のMFキーを格納
for morgan in morgans:
    morgan_keys += morgan.keys()
morgan_keys = list(set(morgan_keys))

morgan_mat = np.zeros((len(morgans), len(morgan_keys))) # 全化合物のMF行列
for idx1, morgan in enumerate(morgans):
    for k, v in morgan.items():
        morgan_mat[idx1][morgan_keys.index(k)] = v
```

```
morgan_mat = pd.DataFrame(morgan_mat)
morgan_mat
```

```
s_bool = morgan_mat == 0.0
s_bool.sum().sum() / (len(morgans) *  len(morgan_keys))
```

```
from sklearn.decomposition import PCA #主成分分析器

#主成分分析の実行
pca = PCA()
pca.fit(morgan_mat)
# データを主成分空間に写像 = 次元圧縮
feature = pca.transform(morgan_mat)
```

```
# 累積寄与率を図示する
import matplotlib.ticker as ticker
import numpy as np
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
plt.plot([0] + list( np.cumsum(pca.explained_variance_ratio_)), "-o")
plt.xlabel("Number of principal components")
plt.ylabel("Cumulative contribution ratio")
plt.grid()
plt.show()
```
